```python
data = {
    'movie' : ['Dune 2', 'KPD4', 'Cars 3'],
    'duration' : ['15 min', '20 min', '45 min']
}
df = pd.DataFrame(data)
df
```

```python
df['duration'].sum()
```

```python
# change time to integers
```

```python
df['duration'] = df['duration'].str.replace('min','').str.strip().astype(int)
df
```
