```python
reg = {
    'reg_id' : [1,2,3,4],
    'name' : ['Andrew','Bob','Claire','David']
}
reg = pd.DataFrame(reg)
reg
```

```python
log = {
    'log_id' : [1,2,3,4],
    'name' : ['Xavier','Andrew','David','Tate']
}
log = pd.DataFrame(log)
log
```

```python
# Inner Join = intersection

pd.merge(reg,log,how='inner',on='name')

```

```python
# Left Join

pd.merge(reg,log,how='left',on='name')
```

```python
# Right Join

pd.merge(reg,log,how='right',on='name')
```

```python
# Full or Outer Join

pd.merge(reg,log,how='outer',on='name')
```

```python
# Sample DataFrames
students = pd.DataFrame({
    'StudentID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David']
})

scores = pd.DataFrame({
    'StudentID': [3, 4, 5, 6],
    'Score': [85, 90, 75, 60]
})

classes = pd.DataFrame({
    'ClassID': [101, 102],
    'ClassName': ['Math', 'English']
})
```

```python
# 1. INNER JOIN
# Keep only students who have scores
pd.merge(students,scores,how='inner',on='StudentID')
```

```python
# 2. LEFT JOIN
# Keep all students, add score if available
pd.merge(students,scores,how='left',on='StudentID')
```

```python
# 3. RIGHT JOIN
# Keep all scores, add student name if available
pd.merge(students,scores,how='right',on='StudentID')
```

```python
# 4. OUTER JOIN
# Keep all students and all scores
pd.merge(students,scores,how='outer',on='StudentID')
```
