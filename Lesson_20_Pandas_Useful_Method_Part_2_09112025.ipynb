Correlatio/Corr()

```python
data = [
    [1691,69],
    [1977,100],
    [1884,109],
    [2151,112],
    [2519,154]
]
myIndex = [2013,2014,2015,2016,2017]
myColumns = ['S&P 500','Apple']
df2 = pd.DataFrame(data=data,index=myIndex,columns=myColumns)
```

# GOOD-PERFECT (0.7 to 1) or (-0.7 to -1) CORRELATION means YOU CAN SEE THE TREND
# 0.8 to 1 - when x increases, y increases (the dots are in a straight line)
# 0.5 to 0.8 - when x increases, y increases (but  the dots are not really in a straight line)
# 0 - no correction (no relationship between x and y)
# -0.5 to -0.8 - when x increases, y decreases (but  the dots are not really in a straight line)
# -0.8 to -1 - when x increases, y decreases (the dots are in a straight line)
