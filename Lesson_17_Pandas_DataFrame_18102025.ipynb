```python
df[(df['total_bill']>= 20) & (df['total_bill']<= 30)]
```

```python
# I want to only get back the rows where size > 5 and its a Thursday
df[(df['size']> 5) & (df['day']=="Thur")]
```

```python
# I want to only get back the rows where size > 5 and it is ANY OTHER DAY OTHER THAN THURSDAY
df[(df['size']> 5) & (df['day']!="Thur")]

```

```python
# I want to only get back the rows where tips is less than 2 and female
df[(df['tip'] < 2) & (df['sex'] == 'Female')]
```

```python
# I want the rows that are weekends ONLY
df[(df['day']=="Sat") | (df['day']=="Sun")]
```

```python
# i want the rows that are weekends ONLY and total_bill which is above the average
```

```python
np.average = df['total_bill'].mean()
df[(df['day']=="Sat") | (df['day']=="Sun") & (df['total_bill'] > np.average)]
```
