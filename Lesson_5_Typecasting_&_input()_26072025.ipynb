# Typecasting

- `converting 1 data type to another data type`

```python
name = "Jun"#String
age = 23 #Integer

print("My name is " + name + "and my age is " + str(age))
```

```python
#String-> Integer : int()
#Integer -> String : str()
#String -> Integer : int()-> String must me numerical string
#Integer -> Float : float()
#Float -> Integer : int()
```

```python
str1 = "9"#Numerical String
str2 = "10"#Numerical String

print(str1+str2)# combine strings
print(int(str1)+int(str2))# typecasting str --> int, then adding them
```

```python
# typecasting from float --> int

num1 = 9.9

print(int(num1))
# when doing typecasting from float --> int, it will always round down
```

```python
num1 = 9.9
num2 = 2.99

print(num1+num2)

print(int(num1)+int(num2))
```

```python
number = 2.7549837598347598739

round(number,2)
```

```python
# strings can be further classified into 2

# (i)  numeric strings     --> "10", "23", "-100", "-2.2"

#Note : Only a numerical string you can convert from a string to float

# (ii) non-numeric strings --> "Jun123", "CodeRangers"

int("Jun123")
```

```python
# You can only typecast on numeric strings into (int/float)

print(int("-10"))
print(type(int("-10")))
```

```python
print(int("12"))
print(type(int("12")))
```

```python
print(float("5.5"))
print(type(float("5.5")))
```

```python
print(float("-45.5"))
print(type(float("-45.5")))
```

# Input()

- `accepting user input`

```python
name = input("Enter your name:")

# the input() command is ASYNCHRONOUS (needs to wait for input BEFORE the below line(s) are run)

print("Hello " + name + ". Welcome to the program!")
```

```python
#ACTUAL CONCLUSION: by default, the input() gives you string data type
# whatever u enter inside the input (even numbers) is a string

num1 = input("Enter the first number: ")
num2 = input("Enter the second number: ")

total = num1 + num2


print("The addition between " + num1 + " and " + num2 + " is "+total)
```

```python
# let's confirm that conclusion

num1 = input("Enter the first number: ")
num2 = input("Enter the second number: ")

print(type(num1))
print(type(num2))
```

```python
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number:"))

total = num1 + num2#now,  the "+" in this case is ADDTION

print("The addition between " + str(num1) + " and " + str(num2) + " is " + str(total) + ".")
```

```python
# there are actually few ways to concatenate in python

# (i)   "+"
# (ii)  ","
# (iii) f strings (formatted string)

name = "Jun"
age = 23

# (i)   "+"
print("Hi my name is "+ name + "." + "My age is " + str(age)+".")

# (ii)  ","
print("Hi my name is ", name, " and I am ", age , "years old.")# with "," you dont need to typecast and manually add spaces

# (iii) f strings (formatted string)

print(f"Hi i am {name} and i am {age} years old.")# with "f strings" you dont need to typecast


# however, it is still advisable for beginners to use "+" for learning typecasting concept purposes
```
