Scattered Plot Basics

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dm_office_sales (1).csv')
df
```

```python
sns.scatterplot(x = 'salary', y = 'sales', data = df)
```

```python
sns.scatterplot(x = 'salary', y = 'division', data = df)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (12, 8))
sns.scatterplot(x = 'salary', y = 'sales', data = df)
```

```python
plt.figure(figsize = (12, 8))
sns.scatterplot(x = 'salary', y = 'sales', data = df, hue = 'division')
```

```python
plt.figure(figsize = (12, 8))
sns.scatterplot(x = 'salary', y = 'sales', data = df, hue = 'work experience') # work experience is a number value so it comes out in one hue
```

```python
plt.figure(figsize = (12, 8))
sns.scatterplot(x = 'salary', y = 'sales', data = df, hue = 'work experience', palette = 'plasma')
```

```python
plt.figure(figsize = (12, 8))
sns.scatterplot(x = 'salary', y = 'sales', data = df, hue = 'work experience', palette = 'inferno')
```

```python
plt.figure(figsize = (12, 8))
sns.scatterplot(x = 'salary', y = 'sales', data = df, s = 200, alpha = 0.2)
```

```python
plt.figure(figsize = (12, 8))
sns.scatterplot(x = 'salary', y = 'sales', data = df, style = 'level of education')
```

```python
plt.figure(figsize = (12, 8))
sns.scatterplot(x = 'salary', y = 'sales', data = df, style = 'level of education', hue = 'level of education', s = 100)
```
