```python
# Exercise 1

#create column called 'after_tax' which implements 10% service tax based on the 'total_bill'
#round to 2 decimal places using np.round()

df['after_tax'] = np.round(df['total_bill'] * 0.9, 2)
df.head()
```

```python
# Exercise 2

#create a new column called 'free_drinks'
#weekends --> true
#weekdays --> false

#df['free_drinks'] = df['day'].apply(lambda day: day in ['Sat', 'Sun'])
#df

def free_drinks(day):
  return day in ['Sat', 'Sun']

df['free_drinks'] = df['day'].apply(free_drinks)
df
```

```python
# Exercise 3

#create a new column called 'Masked CC'
#use apply to mask all but the last four digits of the CC number column(eg: *******0000)

#df['Masked CC'] = df['CC Number'].apply(lambda num: str(num)[:-4] + '*' * 4)

def mask_cc(num):
  num_str = str(num)
  return '*' * (len(num_str) - 4) + num_str[-4:]

df['Masked CC'] = df['CC Number'].apply(mask_cc)
df.head()
```

```python
#using lambda

#Version 1

df['Tip Quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis = 1)
```

```python
#Version 2

def quality(df):
  return df['total_bill']/df['tip']

df['Tip Quality'] = np.round(df.apply(quality, axis = 1), 2)
df.head()
```

```python
# Exercise 1

# Create a new column "Customer Type":

# If total_bill is above $30 and tip is above 20% of total_bill, label as "Premium".
# Otherwise, label as "Regular".

def customer_type(total_bill,tip):
  if total_bill > 30 and (tip/total_bill) > 0.20:
    return 'Premium'
  else:
    return 'Regular'
df['Customer Type'] = df.apply(lambda row: customer_type(row['total_bill'], row['tip']), axis = 1)
df.head()
```

```python
# Exercise 2

# Create a new column "Total Paid" as total_bill + tip

df['Total Paid'] = df['total_bill'] + df['tip']
df.head()
```

```python
# Exercise 3

# Create a new column "Big Tipper":
# If size (number of people) is greater than 4 and tip is greater than 15% of total_bill, return "Yes"; otherwise, "No".

def big_tipper(total_bill,tip,size):
  if size > 4 and (tip/total_bill) > 0.15:
    return 'Yes'
  else:
    return 'No'

df['Big Tipper'] = df.apply(lambda row: big_tipper(row['total_bill'], row['tip'], row['size']), axis = 1)
df.head()
```
