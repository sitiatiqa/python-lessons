String indexing

Accessing a letter of the string
1st letter start with [0][1][2][3]...

```python
word = "Fishball"

print(word[0])
print(word[3])
print(word[-1])
print(word[-3])
```

```python
name = "Siti Atiqa"

print(name[0])
print(name[2])
print(name[4])
print(name[-2])
print(name[-5])
print(name[-1])
```

String slicing

Indexing in a form of word
Accessing more than 1 word

```python
intro = "my name is jong"
print(intro[-5:])
print(intro[0:2])
print(intro[0:1])
print(intro[0:7])
print(intro[-5:7])
print(intro[:-5])
print(intro[7:-5])
```

```python
name = "i am a codeRanger's student"
print(name[0])
print(name[2:4])
print(name[7:11])
print(name[11:17])
```
