# 2d array Indexing

`Indexing - getting 1 item of the array`

*   1D List - Rows of items
*   2D List - Columns & Rows of items

```python
#normal python list

# 1st number - row
# 2nd number - column

list_1d = [1,2,3,4,5,6,7,8,9]
list_2d = [[1,2,3],[4,5,6],[7,8,9]]

print(list_2d[1])
 # row 2 , column 2 --> row index 1, column index 1
print(list_2d[1][1])
```

```python
#LAST ROW FROM THE 2D LIST
print(list_2d[-1])
```

```python
#LAST ELEMENT FROM THE 2D LIST
print(list_2d[-1][-1])
```

```python
list_2d = [[10,20,30],[40,50,60],[70,80,90]] #normal python 2d list

#convert/typecast list --> array
arr_2d = np.array(list_2d)

print(arr_2d) # 2d numpy array
```

```python
# in numpy array, in order to index/slice, u can use the [ , ] notion INSTEAD OF [][]

# 2. [ , ] notation

print(arr_2d[0])
#row 1, column 1
print(arr_2d[1,1])
#row 2, last column
print(arr_2d[2,-1])
print(arr_2d[2,1])
```

```python
# in normal python 2d list, you cant use the [ , ] notation, only [][] notation
print(list_2d[1,1])
```

# 2d array Slicing

`slicing - getting more than 1 item in the array`

*   1D List - Rows of items
*   2D List - Columns & Rows of items

```python
list_2d = [[10,20,30],[40,50,60],[70,80,90]] #normal python 2d list

#convert/typecast list --> array
arr_2d = np.array(list_2d)

print(arr_2d) # 2d numpy array
```

```python
print(arr_2d[1:3])
print('---')
print(arr_2d[1:]) #shortcut
```

```python
print(arr_2d[1:3,1:3])
print('---')
print(arr_2d[1:,1:]) # shortcut
```

```python
print(arr_2d[:,-1]) # all rows, last column
print('---')
print(arr_2d[:,-1].reshape(3,1))
```
