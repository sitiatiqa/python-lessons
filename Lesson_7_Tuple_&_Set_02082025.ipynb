```python
#Set - are not ordered/sequenced, they are randomly shuffled, we use {}
```

```python
list3 = ['orange', 'banana', 'cherry']
print(list3[0])
```

```python
Tuple3 = ['orange', 'banana', 'cherry']
print(Tuple3[0])
```

```python
set1 = {'orange', 'banana', 'cherry'}
print(set1)
```

```python
set2 = set(list3)
print(set2)
```

```python
list4 = ['apple', 'banana', 'apple', 'cherry', 'banana', 'mango']
set4 = set(list4)
print(set4)  # set do not allow duplicates
```

```python
# operation on set
set5 = {'a', 'b', 'c',3}
set6 = {1,2,3,'d'}
```

```python
# operation on set
set5 = {'a', 'b', 'c',3}
set6 = {1,2,3,'d'}
#union
union1 = set5.union(set6)
print(union1)
```

```python
# intersection(), difference(), or symmetric_difference()
```

```python
inter = set5.intersection(set6) #finding common value
```

```python
set5 = {'a', 'b', 'c',3}
set6 = {1,2,3,'d','a'}
inter = set5.intersection(set6)
print(inter)
```
