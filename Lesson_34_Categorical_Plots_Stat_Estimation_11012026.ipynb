```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/dm_office_sales (1).csv")
df
```

Countplot

```python
plt.figure(figsize=(10,4), dpi = 200)
sns.countplot(x = 'division', data = df) # order of x = and data = is pretty important
```

```python
plt.figure(figsize=(10,4), dpi = 200)
sns.countplot(x = 'level of education', data = df)
```

```python
df.groupby('level of education').size().reset_index(name = 'count')
```

```python
plt.figure(figsize=(10,4), dpi = 200)
sns.countplot(x = 'level of education', data = df, palette = 'Set1')
```

```python
plt.figure(figsize=(10,4), dpi = 200)
sns.countplot(x = 'level of education', data = df, hue = 'training level')
```

```python
plt.figure(figsize=(10,4), dpi = 200)
sns.countplot(x = 'level of education', data = df, hue = 'training level', palette = 'Set3')
```

```python
plt.figure(figsize=(10,4), dpi = 200)
sns.countplot(x = 'level of education', data = df, hue = 'training level', palette = 'Paired')
```

Barplot

```python
plt.figure(figsize=(12,6))
sns.barplot(x = 'level of education', y = 'salary',data = df, estimator = np.mean, palette = 'Set1')
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x = 'level of education', y = 'salary',data = df, estimator = np.mean, palette = 'Set1', errorbar = 'sd')
```

```python
stats_df = df.groupby('level of education')['salary'].agg(['mean','std'])
stats_df
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x = 'level of education', y = 'salary',data = df, estimator=  np.mean, hue = 'division')
```
