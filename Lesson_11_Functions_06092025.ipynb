# 1.1 Function Basics

```python
# function --> a box to store code that you can reuse with different inputs
# u can RESUSE THIS BOX
# so u dont need to duplicate the logic

# let's break down the following code:
# (i) 'def' keyword --> define a function
# (ii) 'addition'   --> function's name (customize)
# (iii) (num1,num2) --> function inputs

# step 1: create the box/function
def addition(num1,num2):
  print(num1+num2)# we only created the logic once!

#Step 2 : Use/Reuse box/function
addition(15,5)
addition(12,2)
addition(14,4)
```

## 1.2 Default Inputs (Parameters)

```python
#To solve dynamic input issues, we set a default values for the inputs.

def addition_v3(n1,n2,n3=0):
  print(n1+n2+n3)

addition_v3(1,2)# 2 inputs -->  1 + 2 + 0 (default)
addition_v3(1,2,3)#If you have a value , it will overide the default value
```

```python
def addition_v4(n1=0,n2=0,n3=0):
  print(n1+n2+n3)
```

```python
addition_v4(1,2,3) # n1 = 1, n2 = 2 , n3 = 3
addition_v4(1,2) # n1 = 1, n2 = 2, n3 = 0 (DEFAULT)
addition_v4(1)  # n1 = 1, n2 = 0 (DEFAULT), n3 = 0 (DEFAULT)
addition_v4() # n1 = 0 (DEFAULT), n2 = 0 (DEFAULT), n3 = 0 (DEFAULT)

#This is to prevent crash
```

##1.3 Print vs Return

```python
def multiply(n1,n2):
  print(n1*n2)#The primary function of print is to print the output n1*n2
```

```python
result = multiply(100,10)#It prints the value "1000" but it does not store the value because print function is to print out the output
print(result)#Print none, because "1000" wasnt stored in variable 'result' in the previous line
```

```python
# "return" keyword in function
# when u are using the function, you can choose to store the output in a variable --> return
# Suggestion , when you create a function , by default set it to return rather than print
# To prevent confusion

def multiply(n1,n2):
  return(n1*n2)

result = multiply(2,7) # doesn't print anything because "return" keyword, BUT STORES the value "14" inside the variable 'result'
print('----')
print(result) # prints '14' because '14' was stored in variable 'result' successfully in the previous line due to "return" keyword
```

```python
#create a function called "biggest" and print the biggest number from 3 inputs
# HINT use if else statement
#if (num1 >= num2) AND (num1 >= num3) --> print num1
#if (num2 >= num1) AND (num2 >= num3) --> print num2
#if (num3 >= num1) AND (num3 >= num1) --> print num3


def biggest(n1=0,n2=0,n3=0):
  if(n1>=n2 and n1>=n3):
    return(n1)
  elif(n2>=n1 and n2 >= n3):
    return(n2)
  else:
    return(n3)
```

```python
text = "AAAaaa"
```

```python
print(text.count("a"))#It will return you 3 -> "aaa","AAA" is omited

# problem is that it just gives us the small "a"
```

```python
#Solution we standarize the letters

def count_occurences(text,letter):
  text = text.lower()#Convert the text to lowercase
  letter = letter.lower()#Convert the letter to lowercase
  return(text.count(letter))

print(count_occurences("AAAaaa","a"))
print(count_occurences("BBBbbb","B"))
```
