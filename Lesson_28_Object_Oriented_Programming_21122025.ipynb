```python
class Motorbike:

  # (i) Properties (Charecteristics) - what traits does the motorbike have?
  def __init__(self,name,numberOfWheels,numberOfMirrors):
    self.name = name
    self.numberOfWheels = numberOfWheels
    self.numberOfMirrors = numberOfMirrors

  # (ii) Methods (Functions) - what can the motorbike do?
  def startEngine(self):
    print(self.name + "'s engine started!")

  def honk(self):
    print(self.name + " honked the horn.")

  def brake(self):
    print(self.name + " stepped on the brake!")
```

```python
# creating 2 objects
bike1 = Motorbike("Speedster",6,2)
bike2 = Motorbike("Cruiser",3,1)

# once we created the objects, we can use their CHARACTERISTICS & FUNCTIONS
print(bike1.name)
print(bike1.numberOfWheels)
print(bike1.numberOfMirrors)

bike1.startEngine()
bike1.honk()
bike1.brake()

print('----')

print(bike2.name)
print(bike2.numberOfWheels)
print(bike2.numberOfMirrors)

bike2.startEngine()
bike2.honk()
bike2.brake()
```

## Example 2 - Dog

A `blueprint/recipe (class)` of a Dog:

1. `Characterstics (Properties)`
    - Name given to the dog
    - Breed
    - furColor
    - age
    - hasSibling (if this dog have a sibling(s))

2. `Functions (Methods)`
    - bark

```python
# below is the full code

# advise to code along with students with this following flow:

# 1. create the class with  __init__() --> then straight away create the object and test out properties
# 2. define method (bark) --> then test out
# 3. define method (sleep) --> then test out

# INSTEAD OF DEFINING THE WHOLE CLASS WITH PROPERTIES & METHODS, then only test out
# THIS WILL CONFUSE THE STUDENTS
```

```python
# Part 1- A class is a BLUEPRINT/RECIPE for creating objects
class Dog:

  # (i) Properties (Charecteristics) - what traits does the dog have?

  def __init__(self,name,breed,furColor,age,hasSibling):
    self.name = name
    self.breed = breed
    self.furColor = furColor
    self.age = age
    self.hasSibling = hasSibling

  # (ii) Methods (Functions) - what can the dog do?

  def bark(self):
    print(self.name + " is barking at you!")

  # input 4 --> Spencer is sleeping for 4 hours!
  # no input --> Spencer is sleeping...
  def sleep(self,hours = 0):
    if(hours):
      print(self.name + " is sleeping for "+ str(hours) + " hours!")
    else:
      print(self.name + " is in deep sleep...")
```

```python
# PART 2 - Objects are basically COPIES (can create more than 1 object) created FROM the class/recipe

dog1 = Dog("Spencer", "corgi", "white", 3, True)

print(dog1.name)
print(dog1.hasSibling)

dog1.sleep()
dog1.sleep(4)
```

## Example 3 - BankAccount


```python
# ask students what are some properties and methods of a bank account

# advise to code along with students with this following flow:

# 1. create the class with  __init__() --> then straight away create the object and test out properties
# 2. define method (deposit) --> then test out
# 3. ask student to define method (withdraw) --> they will probably forget u can't withdraw any amount u wish --> if/else needed --> then test out
# 4. define method (display_balance) --> then test out
# 5. ask students what's the logic of (transfer)? how many inputs we need? (amount, obj) --> obj = receiver's bank
# 6. define method (transfer) --> then test out

# INSTEAD OF DEFINING THE WHOLE CLASS WITH PROPERTIES & METHODS, then only test out
# THIS WILL CONFUSE THE STUDENTS
```

```python
# properties (features, what does it have?)
# - account_number
# - balance

# methods (methods, what can it do?)
# - withdraw() "take out money"
# - deposit() "put in money"
# - display_balance()
# - transfer()

class BankAccount:
  def __init__(self, account_number, balance):
    self.account_number = account_number
    self.balance = balance

  def deposit(self, amount):
    self.balance = self.balance + amount
    print("Deposited RM " + str(amount) + ". New balance is RM " + str(self.balance))

  def withdraw(self, amount):
    if(self.balance >= amount):
      self.balance = self.balance - amount
      print("Withdrew RM " + str(amount) + ". New balance is RM " + str(self.balance))
    else:
      print("Insufficient funds. Only have RM " + str(self.balance))

  def display_balance(self):
    print("-" * 30)
    print("Account Number: " + str(self.account_number))
    print("Account Balance: RM " + str(self.balance))
    print("-" * 30)

  def transfer(self,amount,obj):
    if(self.balance >= amount):
      # deduct from this account
      self.balance = self.balance - amount
      # add into the receiver account
      obj.balance = obj.balance + amount

      print("-" * 30)
      print("New balance of this account: RM " + str(self.balance))
      print("New balance of account " + str(obj.account_number) + ": RM " + str(obj.balance))
      print("-" * 30)
```

```python
bank1 = BankAccount(2313586, 1000)
bank2 = BankAccount(2313329, 0)

bank1.display_balance()
bank2.display_balance()

bank1.transfer(400,bank2)
```

## Example 4 - Library Book

```python
# Example: Library Book

# ask students what are some properties and methods of a bank account

# properties (characteristics, what does it have?)
# (ii) title (string)
# (iii) author (string)
# (iv) pages (int)
# (v)  checked_out (boolean)
# (vi) bookmarked_page (int)
# (vii) pages_read (int)

# methods (functions, what can it do?)
# (i) bookmark(page)
# (ii) read(pages)
# (iii) check_out()
# (iv) return_book()
# (v) display_info()
```

```python
# I am pretty sure u check the flow of how to teach (best not to teach all in a go)

# advise to code along with students with this following flow:

# 1. create the class with  __init__() --> then straight away create the object and test out properties
# 2. define method (bookmark) --> ask students can we bookmark any page? --> no --> if/else --> then test out
# 3. define method (display_info) --> then test out
# 4. define method (check_out) --> ask students can we checkout a book if the 'checked_out' property is false? --> no --> if/else --> then test out
# 5. ask student to define method (return_book)
# 6. define method (read) --> explain the logic


# INSTEAD OF DEFINING THE WHOLE CLASS WITH PROPERTIES & METHODS, then only test out
# THIS WILL CONFUSE THE STUDENTS
```

```python
class LibraryBook:
  def __init__ (self,title,author,pages):
    self.title = title
    self.author = author
    self.pages = pages
    self.checked_out = False
    self.bookmark_page = None # empty
    self.pages_read = 0

  def bookmark(self, number):
    if (number <=0 or number > self.pages):
      print("There is no such page. Can't bookmark")
    else:
      self.bookmark_page = number
      print("You have bookmarked page " + str(self.bookmark_page) + "!")

  def display_info(self):
    print('-' * 20)
    print("Book Title: " + self.title)
    print("Book Author: " + self.author)
    print("Page Amount: " + str(self.pages))
    print("Pages Read: " + str(self.pages_read) + "/"  + str(self.pages))
    print("Page Bookmarked: " + str(self.bookmark_page))
    print('-' * 20)

  def check_out(self):
    if(self.checked_out == False):
      self.checked_out = True
      print("Checking out " + self.title + " by " + self.author + ".")
    else:
      print("Cannot check out a book that has already been checked out.")

  def return_book(self):
    if(self.checked_out == False):
      print("Can't return a book that hasn't been checked out.")
    else:
      self.checked_out = False
      print("Returned the book " + self.title + " by " + self.author + ".")


  def read(self, number):
    # if (pages read previously + num of pages u are trying to read) <= book's pages
    if ((self.pages_read + number) > self.pages):
      print("You can't read another " + str(number) + " pages. Only " + str(self.pages - self.pages_read) + " pages remaining." )
    else:
      self.pages_read = self.pages_read + number
      print("You have read " + str(self.pages_read) + " pages. " + str(self.pages - self.pages_read) + " pages remaining.")
```

```python
book1 = LibraryBook("Python 101", "Jun", 500)

book1.display_info()
book1.read(250)
book1.read(400)
```

# Inheritance

Inheritance is a fundamental concept in Object-Oriented Programming (OOP) that `allows a class to inherit properties and methods from another class`.

- The class that inherits is called the
`subclass` or `derived class`,

- and the class from which it inherits is called the `superclass` or `base class`.


Inheritance promotes code reuse and helps in `creating a hierarchical relationship between classes`.

## Example 1

<img src="https://i.ibb.co/ZKDg4R8/Whats-App-Image-2024-06-12-at-18-39-44-3005dcbd.jpg" alt="Whats-App-Image-2024-06-12-at-18-39-44-3005dcbd" border="0" width="450">

```python
# base class /superclass
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return "Some sound"


# Derived class /subclass
class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)  # Call the __init__ method of the Animal class
        self.breed = breed

    def speak(self): # overrides the speak() method in Animal class
        return "Woof!"

    def fetch(self): # this method is specific to dog class
        return f"{self.name} is fetching the ball!"

# Another derived class /subclass
class Cat(Animal):
    # if your subclass doesn't have the '__init()__' method, it automatically inherits the '__init__()' method from its base class/superclass

    def speak(self):
        return "Meow!"#overrides the speak() method in Animal class

    def climb(self, object_to_climb): # this method is specific to dog class
        return f"{self.name} is climbing the {object_to_climb}."

# Create instances of Dog and Cat
dog = Dog("Buddy", "Golden Retriever")
cat = Cat("Whiskers") # uses the '__init()__' of the Animal Class

# Call methods
print(f"{dog.name} says: {dog.speak()}")
print(f"{cat.name} says: {cat.speak()}")

# More method callings
print(dog.fetch())
print(cat.climb("tree"))
```

## Example 2

```python
# Base class / superclass
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def display_info(self):
        return f"Brand: {self.brand}, Model: {self.model}"


# Derived class / subclass: Car
class Car(Vehicle):
    def __init__(self, brand, model, color):
        super().__init__(brand, model)  # Call the __init__ method of the Vehicle class
        self.color = color

    def drive(self):
        return "The car is driving."

    def honk(self):
        return "Beep! Beep!"


# Another derived class / subclass: Motorcycle
class Motorcycle(Vehicle):
    def __init__(self, brand, model, style):
        super().__init__(brand, model)  # Call the __init__ method of the Vehicle class
        self.style = style

    def ride(self):
        return "The motorcycle is riding."

    def rev_engine(self):
        return "Vroom! Vroom!"


# Create instances of Car and Motorcycle
car = Car("Toyota", "Corolla", "Blue")
motorcycle = Motorcycle("Honda", "CBR500R", "Sport")

# Call methods
print(car.display_info())
print(car.drive())
print(car.honk())

print(motorcycle.display_info())
print(motorcycle.ride())
print(motorcycle.rev_engine())
```
