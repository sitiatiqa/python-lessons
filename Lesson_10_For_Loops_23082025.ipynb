`For Loop` acts as a `repeater/iterator` in Python.

It `goes through items that are in a sequence` or `any other iterable/loopable item`.

###1. Looping Through a Range()
Range() is a `loopable/iterable` object, they `contain a sequence of numbers`:

<br>
<b> <u>Example 1: range(0,10)</u></b>

```python
for x in range(0,10):
  print(x)

# range() can have 2 inputs:-> range(staring number:ending number - 1)
# (i)   starting number
# (ii)  ending number (excluded)

#"x" is a variable for your loop that changes valu

#For Example:
# range(0,10) means 0 to 10 (excluding 10)
# 1st Loop              --> "x" = 0 --> print(x) --> print(0)
# 2nd Loop              --> "x" = 1 --> print(x) --> print(1)
# 3rd Loop              --> "x" = 2 --> print(x) --> print(2)
# and so on...
# 10th Loop (Last Loop) --> "x" = 9 --> print(x) --> print(9)
```

<b> <u>Example 2: range(0,21,2)</u></b>

```python
for x in range(0,21,2):
  print(x)

# range() can also have 3 inputs:-> range(starting number, ending number,increment/decrement value )
# (i)   starting number
# (ii)  ending number (excluded)
# (iii) increment/decrement (increase/decrease by how many each loop)

# range(0,21,2) means 0 to 20 (excluding 21),

#"x" is a variable for your loop that changes value DEPENDING ON WHICH LOOP IT IS In

#For Example:
# 1st Loop              --> "x" = 0 --> print(x) --> print(0)
# 2nd Loop              --> "x" = 2 --> print(x) --> print(2)
# 3rd Loop              --> "x" = 4 --> print(x) --> print(4)
# and so on...
# 11th Loop (Last Loop) --> "x" = 20 --> print(x) --> print(20)
```

<b> <u>Example 3: range(100,49,-2)</u></b>

```python
# range(100,49,-2) means 100 to 50 (excluding 49), with STEP SIZE of -2

for x in range(100,49,2):
  print(x)
```

Even strings are `loopable/iterable` objects, they `contain a sequence of characters`:

```python
#Banana

for x in "banana":
  print(x)

#For Example:
# "banana" means b a n a n a --> 6 letters --> 6 loops

# 1st Loop             --> "x" = "b" --> print(x) --> print("b")
# 2nd Loop             --> "x" = "a" --> print(x) --> print("a")
# 3rd Loop             --> "x" = "n" --> print(x) --> print("n")
# and so on...
# 6th Loop (Last Loop) --> "x" = "a" --> print(x) --> print("a")
```

As mentioned before, `string` and `list` are sequence object.

###3. Looping Through a List
Even lists are `loopable/iterable` objects, they `contain a sequence of list items`:


```python
#Loop through the list
grocery_list = ["apple","banana","cherry"]

for x in grocery_list:
  print(x)

#For Example:
# ["apple","banana","cherry] means "apple" "banana" "cherry"

# 1st Loop             --> "x" = "apple"  --> print(x) --> print("apple")
# 2nd Loop             --> "x" = "banana" --> print(x) --> print("banana")
# 3rd Loop (Last Loop) --> "x" = "cherry" --> print(x) --> print("cherry")
```

```python
for x in [123,453,56,343,76,100,21,42]:
  print(x)
```

```python
for x in [123,453,56,343,76,100,21,42]:
  #Filter value > 100 from the list
  if(x>100):
    print(str(x)+ " is more than 100")
  #Filter value < 100 from the list
  elif(x<100):
    print(str(x)+ " is less than 100")
  else:
    print(str(x) + " is exactly 100")
```

###4. Break Keyword

The `break` keyword can be use to `exit the loop`, usually `used when a condition is met`

The `break` keyword is usually used with `if/else` statements inside a ` loop`


```python
list1 = [43,54,32,101,321,532]# For example, if you want to find the 1st number (first occurence) larger than 100 in an array

for x in list1:
  if(x>100):
    print("The first number that is bigger than 100 in this array is " + str(x))
    break

# [43,54,32,101,321,532] --> 6 items --> 6 loops

# 1st loop --> x = 43  --> (is x > 100) --> (is 43 > 100)  --> FALSE --> skip if/else (run next loop)
# 2nd loop --> x = 54  --> (is x > 100) --> (is 54 > 100)  --> FALSE --> skip if/else (run next loop)
# 3rd loop --> x = 32  --> (is x > 100) --> (is 32 > 100)  --> FALSE --> skip if/else (run next loop)
# 4th loop --> x = 101 --> (is x > 100) --> (is 101 > 100) --> TRUE  --> run if/else  -->  "First num > 100 is 101" --> break --> exit loop


# even though supposingly there are 6 loops, we break it at 4th loop (BECAUSE CONDITION IS MET)
```

```python
# Calculate the number of loops it will take to reach the first value "50" in this array

list1 = [43,54,32,50,321,532]

counter = 0 # intialize a "counter" variable (to calculate the number of loops it will take)

for x in list1:
  counter = counter +1
  if(x==50):
    break

print("You need to loop " + str(counter) + " times in this array to get the value of 50.")

# [43,54,32,50,321,532] --> 6 items --> 6 loops

# 1st loop --> x = 43  --> (counter = counter + 1) --> (counter = 0 + 1) --> (counter = 1) --> (is x == 50 ?) --> (is 43 == 50 ?) --> FALSE -->  skip if/else (run next loop)
# 2nd loop --> x = 54  --> (counter = counter + 1) --> (counter = 1 + 1) --> (counter = 2) --> (is x == 50 ?) --> (is 54 == 50 ?) --> FALSE -->  skip if/else (run next loop)
# 3rd loop --> x = 32  --> (counter = counter + 1) --> (counter = 2 + 1) --> (counter = 3) --> (is x == 50 ?) --> (is 32 == 50 ?) --> FALSE -->  skip if/else (run next loop)
# 4th loop --> x = 50  --> (counter = counter + 1) --> (counter = 3 + 1) --> (counter = 4) --> (is x == 50 ?) --> (is 50 == 50 ?) --> TRUE  -->  run if/else --> break --> exit loop

# counter's value is "4" when we break the loop
```

###5. Continue Keyword

The `continue` keyword can be use to `end the current loop and continue to the next loop`, usually `used when a condition is met`

The `continue` keyword is usually used with `if/else` statements inside a ` loop`

```python
for x in range(0,5):
  if x == 2:
    continue # The "continue" statement is used to SKIP THE REMAINING code below, proceed to the next loop
  print("Loop:" + str(x))

# range(0,5) --> [0 1 2 3 4] --> 5 loops

# 1st loop --> x = 0 --> (is x == 2 ?) --> (is 0 == 2) --> False --> skip if/else --> print(Loop: x) --> print(Loop: 0)
# 2nd loop --> x = 1 --> (is x == 2 ?) --> (is 1 == 2) --> False --> skip if/else --> print(Loop: x) --> print(Loop: 1)
# 3rd loop --> x = 2 --> (is x == 2 ?) --> (is 2 == 2) --> True  --> run if/else --> continue (SKIP CURRENT LOOP, dont run code below, go to next loop)
# 4th loop --> x = 3 --> (is x == 2 ?) --> (is 3 == 2) --> False --> skip if/else --> print(Loop: x) --> print(Loop: 3)
# 5th loop --> x = 4 --> (is x == 2 ?) --> (is 4 == 2) --> False --> skip if/else --> print(Loop: x) --> print(Loop: 4)

#When the condition num == 2 becomes True, the "continue" statement gets executed.
#The remaining code (below the "continue", which is the print statement) in the loop is skipped only for that loop.
# Thatâ€™s why "Loop: 5" is missing from the below output.
```

```python
# break    --> exit the WHOLE LOOP
# continue --> exit the CURRENT LOOP
```

### 7. Nested For Loop

Nested For Loop - `a loop inside another loop`

```python
# example 1

# Nested For Loop - a loop inside another loop
# The "inner loop" will be executed one time for each loop of the "outer loop"

adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]


for x in adj:
  for y in fruits:
    print(x + " " + y)

# 1st loop (outer loop - adj) --> x = "red"
#       1st sub loop (inner loop - fruits) --> y = "apple"  --> print(x + " " + y) --> print("red apple")
#       2nd sub loop (inner loop - fruits) --> y = "banana" --> print(x + " " + y) --> print("red banana")
#       3rd sub loop (inner loop - fruits) --> y = "cherry" --> print(x + " " + y) --> print("red cherry")

# 2nd loop (outer loop - adj) --> x = "big"
#       1st sub loop (inner loop - fruits) --> y = "apple"  --> print(x + " " + y) --> print("big apple")
#       2nd sub loop (inner loop - fruits) --> y = "banana" --> print(x + " " + y) --> print("big banana")
#       3rd sub loop (inner loop - fruits) --> y = "cherry" --> print(x + " " + y) --> print("big cherry")

# 3rd loop (outer loop - adj) --> x = "tasty"
#       1st sub loop (inner loop - fruits) --> y = "apple"  --> print(x + " " + y) --> print("tasty apple")
#       2nd sub loop (inner loop - fruits) --> y = "banana" --> print(x + " " + y) --> print("tasty banana")
#       3rd sub loop (inner loop - fruits) --> y = "cherry" --> print(x + " " + y) --> print("tasty cherry")
```

```python
# example 3 - point system

# if 'a'         --> add 10 points
# if 'b'         --> add 5  points
# if 'c'         --> add 3  points
# if 'd' or 'f'  --> add 1  points

sentence = "banana could be my favourite food."
point = 0

for x in sentence:
  if(x.lower() == 'a'):
    point += 10
  elif(x.lower() == 'b'):
    point += 5
  elif(x.lower() == 'c'):
    point += 3
  elif (x.lower() == "d" or x.lower()=='f'):
    point += 1

#'else' means anything other than a,b,c,d,f
print(point)
```

```python
#to determine odd, even numbers in increasing order

numbers = [23,54,1,5,12,65,25,34,4]

odd = sorted([x for x in numbers if x % 2 != 0])
even = sorted([x for x in numbers if x % 2 == 0])

print("Odd List", odd)
print("Even List", even)
```

```python
#to determine odd, even numbers in decreasing order

numbers = [23,54,1,5,12,65,25,34,4]

odd = sorted([x for x in numbers if x % 2 != 0], reverse=True)
even = sorted([x for x in numbers if x % 2 == 0], reverse=True)

print("Odd List", odd)
print("Even List", even)
```
