```python
If/Else statement is an action

Example :
# if it is sunny > go out
# if it is raining > stay inside
```

```python
weather = "Sunny"
if weather == "Sunny":
    print("Go out")
else:
    print("Stay inside")
```

```python
weather = "Sunny"
if weather == "Sunny":
    print("Go out")
elif weather == "Rainy":
    print("Stay inside")
elif weather == "Thundering":
    print("Sleep")
elif weather == "Cloudy":
    print("Bring umbrella")
```

```python
weather = "End of the world"
if weather == "Sunny":
    print("Go out")
elif weather == "Rainy":
    print("Stay inside")
elif weather == "Thundering":
    print("Sleep")
elif weather == "Cloudy":
    print("Bring umbrella")
else:
  print("Error")
```

```python
n = 23
if n % 2 == 1 :
  print("Weird")
elif(n % 2 == 0 and (2 <= n <= 5)):
  print("Not Weird")
elif(n % 2 == 0 and (6 <= n <= 20)):
  print("Weird")
elif(n % 2 == 0 and (n > 20)):
  print("Not Weird")
```

```python
# We can nest the (n % 2 == 0) condition into one statement
```

```python
n = 100
if n % 2 == 0:
  print("Weird")
elif n % 2 == 0:
  if 2 <= n <= 5:
    print("Not Weird")
  elif 6 <= n <= 20:
    print ("Weird")
  elif n > 20:
    print("Not Weird")
```
