## Calculate Wine Consumption Statistics by Continent

### Subtask:
Group the DataFrame by 'continent' and calculate descriptive statistics (count, mean, std, min, 25%, 50%, 75%, max) for 'wine_servings'.


**Reasoning**:
To calculate the descriptive statistics for 'wine_servings' grouped by 'continent', I will use the `groupby()` method on the 'continent' column, select 'wine_servings', and then apply the `describe()` method.



```python
wine_stats = df.groupby('continent')['wine_servings'].describe()
wine_stats
```

```python
df.groupby('continent').mean(numeric_only = True)
```

```python
df.groupby('continent').median.
3(numeric_only = True)
```

**Reasoning**:
The previous subtask of calculating descriptive statistics for 'wine_servings' by continent has been completed. Now, I will calculate the median alcohol consumption for each continent for all relevant alcohol consumption columns (beer_servings, spirit_servings, wine_servings, total_litres_of_pure_alcohol).



```python
median_alcohol_consumption = df.groupby('continent')[['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']].median()
median_alcohol_consumption
```

**Reasoning**:
I need to calculate the mean, minimum, and maximum values for spirit consumption by continent. I will achieve this by grouping the DataFrame by 'continent', selecting the 'spirit_servings' column, and then applying the `agg` function with 'mean', 'min', and 'max'.



```python
spirit_stats_agg = df.groupby('continent')['spirit_servings'].agg(['mean', 'min', 'max'])
spirit_stats_agg
```
