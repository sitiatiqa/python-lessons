```python
list1 = ['bananas', 'apple', 'mango', 'durian']
IDcard = {
    "Name": "Siti", # Name is the key, Siti is the value (key-value pair)
    "Gender" : "Female",
    "Country": "Malaysia"
}
print(list1)
print(IDcard)
print(IDcard["Name"])
print(IDcard["Gender"])
print(IDcard["Country"])
```

```python
# add new key value to Dictionary
IDcard["Age"] = 20
print(IDcard)
```

```python
# update an existing key value
IDcard["Name"] = "Siti Atiqa"
print(IDcard)
```

```python
# remove a key value
del IDcard["Gender"]
print(IDcard)
```

```python
print(IDcard.keys())
print(IDcard.values())
print(IDcard.items())
```
