Rounding Date

```python
ser1 = pd.Series(['2012-12-31 08:45', '2019-1-1 12:30', '2008-02-2 10:30',
               '2010-1-1 09:25', '2019-12-31 00:00'])
custom_idx = ['Day 1', 'Day 2', 'Day 3', 'Day 4', 'Day 5']
ser1.index = custom_idx
ser1
```

```python
ser1 = pd.to_datetime(ser1)
```

```python
ser1.dt.round(freq = 'D')
```

```python
ser1.dt.round(freq = 'H')
```

```python
# Creating timedelta objects

delta_pd1 = pd.Timedelta(days=5)
delta_pd2 = pd.Timedelta(days=1, hours=3, minutes=30)
# Addition
result_pd1 = delta_pd1 + delta_pd2
# Subtraction
result_pd2 = delta_pd1 - delta_pd2
# Comparison
result_pd3 = delta_pd1 == delta_pd2

print(result_pd1)
print(result_pd2)
print(result_pd3)
```
