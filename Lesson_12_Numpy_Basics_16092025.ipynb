## 1.1 There are a few main ways to create `numpy arrays`

1. `array()`
2. `arange()`
3. `zeros and ones`
4. `linspace()`

# 1. array()

```python
list1 = [1,22,333] # normal python list
print(list1)
print(type(list1))
```

```python
# convert/typecast python list --> numpy array


array1 = np.array(list1)

print(array1)

print(type(array1))
```

# 2. arange()

```python
# 2 inputs --> start, stop (excluding stop)

# [0,10)

array2 = np.arange(0,10)

print(array2)
```

```python
list1 = []# creating an empty list

for x in range(0,10):# range(0,10) - 0 1 2 3 4 5 6 8 9 --> 10 numbers --> 10 loops
  list1.append(x)

# "x" is a changing variable depending on which loop it is in

# 1st loop --> x = 0 --> list1.append(x) --> list1.append(0)  --> [0]
# 2nd loop --> x = 1 --> list1.append(x) --> list1.append(1)  --> [0,1]
# 3rd loop --> x = 2 --> list1.append(x) --> list1.append(2)  --> [0,1,2]

print(list1)
```

```python
# 2 inputs --> start, stop (excluding stop)
# 1 input  --> stop (excluding stop), DEFAULTS START TO 0

# if you ONLY provide 1 input in the .arange(), the starting number will DEFAULT to 0

array3 = np.arange(21)

print(array3)
```

```python
# 3 inputs --> start, stop (excluding stop), step
# 2 inputs --> start, stop (excluding stop)
# 1 input  --> stop (excluding stop), DEFAULTS START TO 0

# step size = increment/decrement amount

#--------------------------------------------------
# example of increment (increasing by x)

array4 = np.arange(0,21,2)

print(array4)
```

```python
# creating the same output in NORMAL PYTHON (BEHIND THE SCENES OF NUMPY .arange())

list2 = []#Empty list

for x in range(0,21,2):
  list2.append(x)

print(list2)
```

```python
# example of decrement (decreasing by x)

array5 = np.arange(21,5,-2)

print(array5)
```

```python
# common mistake

array6 = np.arange(21,5,2)
# I am starting from 21, ending at 5, by increasing 2 every time?????

print(array6) # empty array --> THERE IS NO SUCH THING
```

# 4. linspace = linear(evenly) spacing

```python

# I want 21 EVENLY SPACED OUT NUMBERS from 0 to 5
# if you count, there are 21 numbers (21 numbers which are evenly spaced out = increment is the same)
#np.linspace(input1,input2,input3)
#input : Start Range
#Output : Stop Range
#How many numbers to return

array10 = np.linspace(0,5,21)

print(array10)
```

```python
# another example

# I want 15 EVENLY SPACED OUT NUMBERS from 5 to 20

arr11 = np.linspace(5,20, 15)

print(arr11)
print("---"*10)
print(len(arr11))
```

## 1.2 Random numbers in numpy arrays

There are different ways to generate random numbers in a numpy array

1. `rand()`
2. `randint()`
3. `seed()`

```python
#Generating 5 random numbers from 0-1

# when we are dealing with random numbers, we have to use a SUBLIBRARY/SUBPACKAGE in numpy --> (np.random)

# library                   - main chapter - Chapter 1          -  (numpy)
# sublibrary                - subchapter - Chapter 1.1          -  (random)
# method/commands           - topic    - Chapter 1.1.1 (hello)  -     (rand)

arr1 = np.random.rand(5)

print(arr1)
```

```python
# create a matrix of rand()

# matrix (mathematical term) - rows & columns


arr2 = np.random.rand(4,2)

print(arr2)
```

# 2. randint


```python
#np.random.randint(start value , stop value )

num1 = np.random.randint(0,100)

print(num1)
```

```python
# 3 inputs

# 1st input - start
# 2nd input - end (exclude end)
# 3rd input - how many items in the array?


arr3 = np.random.randint(50,100,20)

print(arr3)
```

# 3. seed

```python
np.random.rand(4)
```

```python
np.random.seed(101)
np.random.rand(4)
```

## 1.4 Dtype

```python
arr1 = np.array([11,22,33,44])

print(type(arr1)) #  data type --> numpy array
print("-"*10)
print(arr1.dtype) # dtype --> int (specifically refers to the data type of the elements stored in the array)
```

```python
arr1 = np.array([2.2,3.3,4.4,5.5])

print(type(arr1)) #  data type --> numpy array
print('---')
print(arr1.dtype) # dtype --> float (specifically refers to the data type of the elements stored in the array)
```

```python
arr1 = np.array(["jun","jun2","jun3"])

print(type(arr1)) #  data type --> numpy array
print('---')
print(arr1.dtype) # dtype --> <U4 (string) (specifically refers to the data type of the elements stored in the array)
# <U4 means strings under 4 characters long

# unicode
# encoding  (changing from characters to binaries)
# a --> 010
```
