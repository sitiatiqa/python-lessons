```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/StudentsPerformance (1).csv')
df.head()
```

Joint Plot

```python
sns.jointplot(x='math score', y='reading score', data=df)
```

```python
sns.jointplot(x='math score', y='reading score', data=df, kind='hex') # kind is where you define your layout
```

```python
sns.jointplot(x='math score', y='reading score', data=df, kind='hist')
```

```python
sns.jointplot(x='math score', y='reading score', data=df, kind='kde')
```

```python
sns.jointplot(x='math score', y='reading score', data=df, kind='resid')
```

```python
sns.jointplot(x='math score', y='reading score', data=df, kind='reg')
```

Pair Plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='gender', palette = 'viridis')
```

```python
sns.pairplot(df, hue='gender', palette = 'viridis', diag_kind='hist')
```

```python
sns.pairplot(df, hue='gender', palette = 'viridis', corner = True)
```
