```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/dm_office_sales (1).csv")
df
```

```python
df.info()
```

```python
df['division'].unique()
```

1. Rugplot - a simple plot that displays individual data points along a single axis, usually the x-axis as small lines or ticks
- continuous feature, small range


```python
sns.rugplot(x = 'salary', data = df) # y-axis doesn't represent anything
```

```python
sns.rugplot(x = 'division', data = df)
```

2. Displot - a figure-level function that draws histograms, kernel density estimates, or empirical cumulative distribution functions for univariate or bivariate data
- the function allows you to map to a figure, rather than an axes object


```python
sns.displot(data = df, x = 'salary')
```

```python
sns.histplot(data = df, x = 'salary') # same as displot, can add another feature to make graph more informatic
```

```python
sns.histplot(data = df, x = 'salary', kde = True) # kernel density estimator
df['salary'].mean()
```

```python
from matplotlib.lines import lineStyles
sns.histplot(data = df, x = "salary")
mean_value = np.mean(df['salary'])

plt.axvline(mean_value, color='red', linestyle='--')

# text() needs 3 inputs (x,y,text)
plt.text(mean_value * 1.05 , 100, f'Mean: {mean_value:.2f}', color='red')

plt.title("Histogram with Mean Line")
plt.show()
```

```python
sns.histplot(data = df, x = 'salary', bins = 10) # bins = bar, crucial for accuracy
```

```python
sns.histplot(data = df, x = 'salary', bins = 100)
```

```python
sns.set_theme(style = 'darkgrid')
sns.histplot(data = df, x = 'salary', bins = 100)
```

```python
sns.displot(data = df, x = 'salary', bins = 20, color = 'red', lw = 6, edgecolor = 'black', ls = '--')
```
