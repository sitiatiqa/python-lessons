```python
import numpy as np
import matplotlib.pyplot as plt
```

Basic Example

```python
x = np.arange(0,10)
x
```

```python
y = np.arange(0,20,2)
y
```

```python
plt.plot(x,y)
plt.xlabel('X Axis Title here')
plt.ylabel('Y Axis Title here')
plt.title('String Title here')
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('X Axis Title here')
plt.ylabel('Y Axis Title here')
plt.title('String Title here')
plt.xlim(0,6)
plt.ylim(0,12)
plt.show()
```
