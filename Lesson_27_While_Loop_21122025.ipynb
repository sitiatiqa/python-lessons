# While Loop vs For Loop



````python
# for loop   --> runs as long as it's in a range (number, string, list)
# while loop --> runs as long as the condition is true

# Even though the definitions are clear, many students still get confused about when to use a while loop and when to use a for loop.

# Here are some tips on when to use each:

# If you need to go through a list, string, or range of numbers ONCE, use a for loop.
# If you need to repeat something UNTIL a certain condition is met (and you don’t know exactly how many times it will run), use a while loop.
````

# Use Case 1

````python
# ATM PIN ENTRY

# ask students do we use while or for loop
````

````python
# answer: For Loop

# because if u use while loop, u can technically have INFINITE PASSWORD ATTEMPTS until you logged in (stop loop)
````

````python
# while loop --> not a good solution (infinite attempts)
user_input = ""
password = "garry123"

while user_input != password:
  user_input = input("Enter your pin: ")
  if user_input != password:
    print("Incorrect password. Try again. ")

print("Sucessful login")
````

````python
# for loop --> good solution (finite attempts)

password = "garry123"

for x in range(0,3):
  user_input = input("Enter your pin: ")
  if(user_input == password):
    print("Successful login.")
    break
  else:
    print(f"Wrong password")
else:
  # this 'else' is for the 'for' loops (when it ends)
  print("No attempts left")

````

````python
# version 2 of the for loop --> showing users number of attempts left

password = "garry123"

for x in range(0,3):
  user_input = input("Enter your pin: ")
  if(user_input == password):
    print("Successful login.")
    break
  else:
    print(f"Wrong password. Attempts left {2-x}")  # x = range(0,3) --> 0 1 2
else:
  print("No attempts left")


````

# Use Case 3


````python
#  in a game, u have 3 lives with multiple rounds, game continues until player loses all 3 lives

lives = 3

# ask students do we use while or for loop
````

````python
# answer --> WHILE Loop

# reason why it is not FOR, ur live dont decrease every round (depends if u won/lost) --> "condition"


lives = 3

while lives > 0:
  print(f"You have {lives} lives left.")
  action = input("Did you win or lose the round (yes/no): ")
  if(action == "no"):
    lives -=1

print("Game over")
````

# Use Case 6

````python
# food order system

menu = {
    "coffee": 3.5,
    "tea": 2.5,
    "cake": 4.0,
    "sandwich": 5.0,
    "cookie": 1.5
}

total = 0

# ask students do we use while or for loop
````

````python
# Answer --> While Loop, because its an order system, u dont need to go through all items once

# sometimes u just want coffee, and sandwich, without ordering other items
# sometimes u might have multiple orders of the same food item

menu = {
    "coffee": 3.5,
    "tea": 2.5,
    "cake": 4.0,
    "sandwich": 5.0,
    "cookie": 1.5
}

total = 0

while True:
  order  = input("Enter item ('done' to finish): ")
  if order == "done":
    break
  elif order in menu:
    total += menu[order]
    print(f"Added {order} = RM{menu[order]}")
    print(f"Current total: RM{total}")
    print("-"*50)
  else:
    print("Item not found")
    print("-"*50)



print(f"Total bill: RM{total}")
````

````python
# version 2 of the program

# the previous version has a problem
# notice how i wanted to order 3 cakes

# i had to enter cake 3 times! What if when u wanted to order a food item,
# the program prompts you the amount you want as well

menu = {
    "coffee": 3.5,
    "tea": 2.5,
    "cake": 4.0,
    "sandwich": 5.0,
    "cookie": 1.5
}

total = 0

# infinite loop
while True:
  order  = input("Enter item ('done' to finish): ")
  if order == "done":
    break
  elif order in menu:
    amount = int(input(f"Enter the amount for {order}: "))
    itemTotal = amount * menu[order]
    total += itemTotal
    print(f"Added {amount} {order} = RM{itemTotal}")
    print(f"Current total: RM{total}")
    print("-"*50)
  else:
    print("Item not found")
    print("-"*50)


print(f"Total bill: RM{total}")
````

# Use Case 7





````python
# Sum of the first 100 natural numbers

total = 0

# ask students do we use while or for loop
````

````python
# Answer --> For Loop/You know exactly how many times the loop needs to run (from 1 to 100).
# It’s a fixed range, not a condition that changes unpredictably.

total = 0
for i in range(1, 101):
    total += i
print(total)
````
